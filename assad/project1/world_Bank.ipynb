```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

m = pd.read_excel(r"C:\Users\SONY\Desktop\bootcamp\project1\proj4.xls")
mpl.rcParams['lines.linestyle'] = '--'
mpl.rcParams['lines.linewidth'] =3
x = m['year']
agriculture = m['Employment in agriculture (% of total employment) (modeled ILO estimate)']
industry = m['Employment in industry (% of total employment) (modeled ILO estimate)']
def demo(sty):
    mpl.style.use(sty)
    fig, ax = plt.subplots(figsize=(15, 5))
    ax.plot(x,agriculture,'C10',label='agriculture')
    ax.plot(x,industry,label='industry')
    ax.set_title(' Comparison of total employment between agriculture and industry'.format(sty), color="k")
    plt.xlabel('Year')
    plt.ylabel('Employment in agriculture & industry')
    plt.legend()
demo('seaborn')
```
