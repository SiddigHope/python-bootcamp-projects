```python
th = np.linspace(0, 2*np.pi, 128)

def demo(sty):
    mpl.style.use(sty)
    fig, ax = plt.subplots(figsize=(12,3))

#     ax.set_title('style: {!r}'.format(sty), color='C0')

    ax.plot( a, 'C1', label='Military expenditure (% of GDP)')
    ax.plot( b, 'C2', label='Health expenditure (% of GDP)')
    ax.plot( c, 'C3', label='Education expenditure(% of GDP)')
    txt="Figure: Ilustrates Government Expenditures in three Sectors from GDP "
    plt.figtext(0.5,1.0, txt, wrap=True, horizontalalignment='center', fontsize=14)
#     ax.plot('x','y', pop_65[55:], 'Populatin for Age 65 - 69')
#     ax.plot('x','y', pop_14[55:], 'Populatin for Age 0 - 14')
    ax.legend()
    

demo('default')
```
